````python
import json, os
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output
````

````python
patient_ids = sorted([f.replace('.json','') for f in os.listdir('outputs') if f.endswith('.json')])

selector = widgets.Dropdown(options=patient_ids, description='Select Patient:')
out = widgets.Output()

def on_change(change):
    with out:
        clear_output()
        pid = change['new']
        with open(f'outputs/{pid}.json') as f:
            note = json.load(f)
        display(Markdown(f"## Clinical Note – Patient {pid}"))
        display(Markdown(f"```json\n{json.dumps(note, indent=2)}\n```"))

selector.observe(on_change, names='value')
display(selector, out)
on_change({'new': selector.value})
````
